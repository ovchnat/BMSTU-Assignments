<p>5. Постройте диаграммы разброса (Ваша компания - Компания с min), (Ваша компания - Компания с max), (Ваша компания - Компания с none)</p>

```python
plt.figure(figsize=[18, 5])

plt.subplot(1, 3, 1)

plt.scatter(x = st_df["BABA"],y = st_df[alb_df.idxmin()], color = 'fuchsia')
plt.xlabel("BABA",size = 10)
plt.ylabel(alb_df.idxmin(), size = 10)
plt.grid(True)

plt.subplot(1, 3, 2)
plt.scatter(x = st_df["BABA"],y = st_df[alb_df.idxmax()], color = 'indigo')
plt.xlabel("BABA",size = 15)
plt.ylabel(alb_df.idxmax(),size = 15)
plt.grid(True)

plt.subplot(1, 3, 3)
plt.scatter(x = st_df["BABA"],y = st_df[abs(alb_df).idxmin()], color = 'purple')
plt.xlabel("BABA",size = 15)
plt.ylabel(abs(alb_df).idxmin(), size = 15)
plt.grid(True)

plt.show()
```
